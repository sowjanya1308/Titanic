```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
gender_submission_df = pd.read_csv('gender_submission.csv')
gender_submission_df.head(10)
```

```python
train_df = pd.read_csv('train.csv')
train_df.head(10)
```

```python
test_df  = pd.read_csv('test.csv')
test_df.head(10)
```

```python
train_df['Age'].fillna(int(train_df['Age'].mean()),inplace=True)
```

```python
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0],inplace=True)
```

```python
train_df.isnull().sum()
```

```python
train_df = train_df.drop(columns=['Cabin'])
```

```python
train_df.isnull().sum()
```

```python
categorical_columns_train = train_df.select_dtypes(include=['object']).columns
print("Categorical columns in the training dataset:")
print(categorical_columns_train)
```

```python
categorical_columns_train = ['Sex', 'Embarked']
for col in categorical_columns_train:
    print(f"\n{col}: {train_df[col].unique()}")
```

```python
encoding_data = pd.get_dummies(train_df,columns=['Sex','Embarked'],dtype=int)
```

```python
encoding_data.head()
```

```python
encoding_data = encoding_data.drop(['Name','Ticket','PassengerId'],axis=1)
```

```python
encoding_data.head()
```

```python
X = np.asarray(encoding_data.drop('Survived',axis=1))
sc = StandardScaler()
X = sc.fit_transform(X)
y = np.asarray(encoding_data['Survived'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
def classification_model():
    model = Sequential()
    model.add(Dense(100, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(25, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))  # Correct activation function for binary classification
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])  # Correct loss function for binary classification
    return model
```

```python
# Build the model
model = classification_model()

# Fit the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)

# Evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)

# Print the final accuracy
print(f'Final Accuracy: {scores[1] * 100:.2f}%')
```

```python
new_data = pd.DataFrame({
    "Pclass": [3],
    "Age": [25],
    "SibSp": [1],
    "Parch": [0],
    "Fare": [7.25],
    "Sex_female": [0],
    "Sex_male": [1],
    "Embarked_C": [0],
    "Embarked_Q": [0],
    "Embarked_S": [1]
})

# Preprocess the new input data
X_new = sc.transform(new_data)

# Make predictions on the new data
new_predictions = model.predict(X_new)

# Convert probabilities to class labels (0 or 1)
new_predicted_classes = (new_predictions > 0.5).astype(int)

# Print the predictions for the new data
print("Predicted classes for new data:")
print(new_predicted_classes)
```
